```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Loading the data
df = pd.read_csv('data/ReportSample.csv', low_memory=False)
```

```python
# Group by Type and get TID and Unique IP count
type_summary = df.groupby('Category').agg({
    'TID': 'count',
    'IP': pd.Series.nunique
}).rename(columns={'TID': 'Count of TID', 'IP': 'Count of IP'})

# Get top 3 by Count of TID
top3_types = type_summary.sort_values('Count of TID', ascending=False).head(3)

# Plot
ax = top3_types.plot(kind='bar', figsize=(10,6))
plt.title("Top 3 Threat Categories")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()

# Adding value labels to each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

plt.show()
```

```python
# Group by Type and get unique IPs
type_summary = df.groupby('Category')['IP'].nunique()

# Get Top 3
top3_ip_types = type_summary.sort_values(ascending=False).head(3)

# Doughnut chart
plt.figure(figsize=(3,3))
plt.pie(top3_ip_types, labels=top3_ip_types.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
plt.title("Top 3 Threat Categories by Unique IPs")
plt.show()
```

```python
# Filter the vulnerability rows
vuln_df = df[df['Category'] == 'Vulnerability']

# Create summary values
frequency = vuln_df['TID'].count()
unique_ips = vuln_df['IP'].nunique()

# Build single-row table as DataFrame
vuln_summary = pd.DataFrame({
    'Vulnerability Observed': ['Vulnerability'],
    'Frequency of Occurrence': [frequency],
    'Count of Unique IP': [unique_ips]
})

# Display the result
vuln_summary
```

```python
# Group, count, and sort
region_summary = (
    df.groupby('Region')
      .agg({'TID': 'count', 'IP': pd.Series.nunique})
      .rename(columns={'TID': 'Count of TID', 'IP': 'Count of Unique IP'})
      .sort_values(by='Count of TID', ascending=False)
      .reset_index()
)

# Add total row
total = pd.DataFrame({
    'Region': ['Total'],
    'Count of TID': [region_summary['Count of TID'].sum()],
    'Count of Unique IP': [region_summary['Count of Unique IP'].sum()]
})

# Combine
final_table = pd.concat([region_summary, total], ignore_index=True)

# Display result
final_table
```

```python
# Group by Region
region_summary = df.groupby('City').agg({
    'TID': 'count',
    'IP': pd.Series.nunique
}).rename(columns={'TID': 'Count of TID', 'IP': 'Count of IP'})

# Get Top 5
top5_states = region_summary.sort_values('Count of TID', ascending=False).head(5)

# Plot
ax = top5_states.plot(kind='bar', figsize=(10,6), color=['steelblue', 'orange'])
plt.title("Top 5 States with Most Threats")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()

# Adding value labels to each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

plt.show()
```

```python
# Group by CC_Ip and count unique IPs
ccip_summary = df.groupby('CC_Ip')['IP'].nunique()

# Get Top 5
top5_ccip = ccip_summary.sort_values(ascending=False).head(5)

# Doughnut chart
plt.figure(figsize=(6,3))
plt.pie(top5_ccip, labels=top5_ccip.index, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.4})
plt.title("Top 5 C&C IPs by Unique IPs")
plt.show()
```

```python
# Filter for Malware and SpamBot types
malware_df = df[df['Type'].isin(['Malware', 'SpamBot'])]

# Group by Infection
malware_summary = malware_df.groupby('Infection').agg({
    'TID': 'count',
    'IP': pd.Series.nunique
}).rename(columns={
    'TID': 'Frequency of Occurrence',
    'IP': 'Count of Unique IPs'
}).reset_index()

# Sort and select top 5
top5_malware = malware_summary.sort_values(by='Frequency of Occurrence', ascending=False).head(5)

# Set Infection as index for plotting
ax = top5_malware.set_index('Infection')[['Frequency of Occurrence', 'Count of Unique IPs']].plot(
    kind='bar',
    figsize=(10,6),
    color=['steelblue', 'orange']
)

# Styling
plt.title("Top 5 Prevalent Malware Observed (Including SpamBot)")
plt.ylabel("Count")
plt.xlabel("Infection")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding value labels to each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

# Show plot
plt.show()
```
